```python
from numpy import *
import matplotlib.pyplot as plt 

def Ffun(x, y, R, Q):
	r = sqrt(x**2+y**2)
	F = where(r <= R, Q**2/(r**2), sqrt(Q)/r)
	return (r, F)

R = 20.0
x = linspace(R, 4*R, 101).reshape(-1,1)
y = arange(R, 4*R, 0.08*R).reshape(1,-1)

Q = 2
r, F = Ffun(x, y, R, Q)#101,38


f = plt.figure()

ax1 = plt.subplot(2,1,1)
for i in range(r.shape[1]):
	cx = r[10, i]
	cy = F[10, i]
	print(cy)
	ax1.plot(cx, cy)
	
	
ax1.set_xlabel('r')
ax1.set_ylabel('F')
ax1.set_title('fila 10')

ax2 = plt.subplot(2,1,2)
for i in range(0, r.shape[1], 1):
	cx = r[-1, i]
	cy = F[-1, i]
	ax2.plot(cx, cy)
	
	
ax2.set_xlabel('r')
ax2.set_ylabel('F')
ax2.set_title('ultima fila')

plt.show()

```
