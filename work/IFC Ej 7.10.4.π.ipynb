```python
from numpy import *

import matplotlib.pyplot as plt
```

```python
h = arange(10, 110, 10)
g = array([ 9.76, 9.74, 9.70, 9.69, 9.66, 9.62, 9.59, 9.55, 9.54, 9.51 ])

print(f"h = {h}")
print(f"g = {g}")
```

```python
# Ver: https://queirozf.com/entries/add-labels-and-text-to-matplotlib-plots-annotation-examples
fig = plt.figure()
plt.scatter(h, g)
for x, y in zip(h, g):
    label = f"({x},{y})"
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.xlim(0, 110)
plt.ylim(9, 10)
plt.show()
```

```python
# Ver: https://www.geeksforgeeks.org/linear-regression-python-implementation/

def estimate_coef(x, y):
  # number of observations/points
  n = size(x)
 
  # mean of x and y vector
  m_x = mean(x)
  m_y = mean(y)
 
  # calculating cross-deviation and deviation about x
  SS_xy = sum(y*x) - n*m_y*m_x
  SS_xx = sum(x*x) - n*m_x*m_x
 
  # calculating regression coefficients
  b_1 = SS_xy / SS_xx
  b_0 = m_y - b_1*m_x
 
  return (b_0, b_1)

h0, a = estimate_coef(h, g)
print(f"a = {a}")
print(f"h0 = {h0}")
```

```python
# Ver: https://queirozf.com/entries/add-labels-and-text-to-matplotlib-plots-annotation-examples
fig = plt.figure()
j = array([ (h0 + a * hh) for hh in h ])
plt.scatter(h, j)
for x, y in zip(h, j):
    label = f"({x},{y:.02f})"
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.xlim(0, 110)
plt.ylim(9, 10)
plt.show()
```

```python
# Ver: https://es.wikipedia.org/wiki/Regresi%C3%B3n_lineal

print(f"h = {h}")
print(f"g = {g}")

n = size(h)
print(f"n = {n}")

h_mean = mean(h)
g_mean = mean(g)
print(f"h_mean = {h_mean}")
print(f"g_mean = {g_mean}")

ngh = n * g_mean * h_mean
print(f"ngh = {ngh}")

sum_gh = sum(g * h)
print(f"sum_gh = {sum_gh}")
sum_h2 = sum(h * h)
print(f"sum_h2 = {sum_h2}")
```

```python
a = (sum_gh - ngh) / ( sum_h2 - (1 / n) * (sum(h) ** 2) )
print(f"a = {a}")
h0 = g_mean - a * h_mean
print(f"h0 = {h0}")
```

```python
# Ver: https://queirozf.com/entries/add-labels-and-text-to-matplotlib-plots-annotation-examples
fig = plt.figure()
j = array([ (h0 + a * hh) for hh in h ])
plt.scatter(h, g, color='blue')
plt.scatter(h, j, color='red')
plt.xlim(0, 110)
plt.ylim(9.4, 9.8)
plt.show()
```
