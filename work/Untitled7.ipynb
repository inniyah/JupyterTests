```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
def coordenadas (v, a, t, g=9.81):
	x = v*t*cos(a)
	y = v*t*sin(a) - 0.5*g*t**2
	x = expand_dims(x, axis=-1)
	y = expand_dims(y, axis=-1)
	return concatenate([x, y], axis=-1)
```

```python
v = arange(50.0, 100.0, 10.0).reshape(-1,1,1)
a = arange(45.0, 90.0, 5.0).reshape(1,-1,1)
t = linspace(0.0, 10.0, 50).reshape(1,1,-1)
```

```python
resul = coordenadas(v, a*pi/180, t)#5,9,50,2
print(v.shape)
print(a.shape)
print(t.shape)

print(resul.shape)
```
