```python
from numpy import *

import matplotlib.pyplot as plt
from matplotlib import cm

# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Creamos un mayado homogneo en el plano XY (forma alternativa)
xv = yv = arange(-5, 0, 0.1)
x, y = meshgrid(xv, yv);
# Calculamos las coordenada de la superficie
# en cada uno de los puntos (x,y)
z = 0.2*sin(x)*exp(-(x+3)**2)*exp(-(y+2)**2)

# Creamos la figura y los ejes 3D
fig = plt.figure()
# Ver: https://matplotlib.org/2.0.2/mpl_toolkits/mplot3d/tutorial.html
ax = fig.add_subplot(111, projection='3d')
# Pintamos la superficie
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)
# Etiquetamos los ejes
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
# Mostramos la figura
plt.show()
```

```python
# Ver: https://scipy-lectures.org/intro/matplotlib/auto_examples/options/plot_colormaps.html

import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=False)
a = np.outer(np.arange(0, 1, 0.01), np.ones(10))

plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=0.8, bottom=0.05, left=0.01, right=0.99)
maps = [m for m in plt.cm.datad if not m.endswith("_r")]
maps.sort()
l = len(maps) + 1

for i, m in enumerate(maps):
    plt.subplot(1, l, i+1)
    plt.axis("off")
    plt.imshow(a, aspect='auto', cmap=plt.get_cmap(m), origin="lower")
    plt.title(m, rotation=90, fontsize=10, va='bottom')

plt.show()


# https://scipy-cookbook.readthedocs.io/items/Matplotlib_Show_colormaps.html

# https://matplotlib.org/stable/users/explain/colors/colormaps.html

# https://www.fieldtriptoolbox.org/faq/colormap/

# https://blog.csdn.net/MizarTian/article/details/88664412

# https://foundations.projectpythia.org/_images/sequential.png
```

```python
# Creamos la figura
fig = plt.figure()

# Ver: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html
# Ver: https://matplotlib.org/stable/gallery/mplot3d/subplot3d.html


# Arriba, a la izquierda
ax = fig.add_subplot(2, 2, 1, projection='3d')

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

# Creamos los datos
# Ver: Ejemplo 8.4.3

# Creamos un mayado homogneo en el plano XY (forma alternativa)
xv = yv = arange(-5, 0, 0.1)
x, y = meshgrid(xv, yv);

# Calculamos las coordenada de la superficie
# en cada uno de los puntos (x,y)
z = 0.2*sin(x)*exp(-(x+3)**2)*exp(-(y+2)**2)

# Pintamos la superficie
# Ver: https://matplotlib.org/stable/api/_as_gen/mpl_toolkits.mplot3d.axes3d.Axes3D.plot_surface.html

# rstride, cstride: Downsampling stride in each direction
# cmap: Colormap of the surface patches
p = ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.jet)

# Ver: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xticks.html
ax.set_xticks(arange(-5, 1, 1))
ax.set_yticks(arange(-5, 1, 1))
#ax.set_zticks([-0.05, 0, 0.05])

# Ver: https://matplotlib.org/stable/users/explain/axes/colorbar_placement.html
plt.colorbar(p, ax=ax)


# Arriba, a la derecha
ax = fig.add_subplot(2, 2, 2 , projection='3d')

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

yv = zv = arange(-5, 0, 0.1)
y, z = meshgrid(yv, zv);
x = 0.2*sin(y)*exp(-(y+3)**2)*exp(-(z+2)**2)

p = ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.YlOrBr.reversed())

#ax.set_xticks(arange(-0.10, 0.10, 0.05))
ax.set_yticks(arange(-5, 1, 1))
ax.set_zticks(arange(-5, 1, 1))

plt.colorbar(p, ax=ax)


# Abajo, a la izquierda
ax = fig.add_subplot(2, 2, 3 , projection='3d')

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

xv = zv = arange(-5, 0, 0.1)
x, z = meshgrid(xv, zv);
y = 0.2*sin(x)*exp(-(x+3)**2)*exp(-(z+2)**2)

p = ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.cool)

ax.set_xticks(arange(-5, 1, 1))
#ax.set_yticks(arange(-0.10, 0.10, 0.05))
ax.set_zticks(arange(-5, 1, 1))

plt.colorbar(p, ax=ax)


# Abajo, a la derecha
ax = fig.add_subplot(2, 2, 4 , projection='3d')

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

xv = yv = arange(-5, 0, 0.1)
x, y = meshgrid(xv, yv);
z = 0.2*sin(x)*exp(-(x+3)**2)*exp(-(y+2)**2)

p = ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.Greens)

ax.set_xticks(arange(-5, 1, 1))
ax.set_yticks(arange(-5, 1, 1))
#ax.set_zticks([-0.05, 0, 0.05])

plt.colorbar(p, ax=ax)


# Mostramos la figura
plt.show()
```
