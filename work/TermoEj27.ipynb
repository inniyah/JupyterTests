```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las ecuaciones del sistema
def system(x, y):
    x_dot = x - x * y
    y_dot = y - x * y
    return x_dot, y_dot

# Crear una malla de puntos para graficar el campo vectorial
x = np.linspace(-0.5, 1.5, 20)  # Rango para x
y = np.linspace(-0.5, 1.5, 20)  # Rango para y
X, Y = np.meshgrid(x, y)

# Calcular el campo vectorial
U, V = system(X, Y)

# Normalizar el campo vectorial para graficar
magnitude = np.sqrt(U**2 + V**2)
U_norm = U / (magnitude + 1e-8)  # Evitar divisin por cero
V_norm = V / (magnitude + 1e-8)

# Graficar el campo vectorial
plt.figure(figsize=(8, 8))
plt.quiver(X, Y, U_norm, V_norm, color='blue', alpha=0.7)  # Flechas del campo

# Agregar los puntos crticos
critical_points = [(0, 0), (1, 1)]
for cp in critical_points:
    plt.plot(cp[0], cp[1], 'ro', markersize=10, label=f"Punto crtico {cp}")

# Configuracin de la grfica
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Eje x
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Eje y
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.title('Campo vectorial del sistema dinmico', fontsize=16)
plt.legend()
plt.grid(alpha=0.3)

# Mostrar la grfica
plt.show()
```
