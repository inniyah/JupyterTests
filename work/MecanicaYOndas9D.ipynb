```python
import sympy as sp
import numpy as np

from sympy.solvers.ode import dsolve

from scipy.integrate import solve_ivp

import matplotlib.pyplot as plt

from pprint import pprint as ppr
```

```python
t = sp.Symbol('t')
vy = sp.Function('vy')(t)
vx = sp.Function('vx')(t)

v0 = 1
q = 1
m = 1
E = 1
B = 1

ec1 = vx.diff(t) - q / m * ( E + B * vy )
ec2 = vy.diff(t) + q / m * ( B * vx )

sol = dsolve([ec1, ec2], [vx, vy], ics={vx.subs(t, 0):v0, vy.subs(t, 0):0})

sp.pprint([s.simplify() for s in sol])

plt.figure()
sp.plot(sol[0].rhs,
        sol[1].rhs)
```

```python
t = sp.Symbol('t')
v0 = sp.Symbol('v0')
q = sp.Symbol('q')
m = sp.Symbol('m')
E = sp.Symbol('E')
B = sp.Symbol('B')
vy = sp.Function('vy')(t)
vx = sp.Function('vx')(t)

ec1 = vx.diff(t) - q / m * ( E + B * vy )
ec2 = vy.diff(t) + q / m * ( B * vx )

sol = dsolve([ec1, ec2], [vx, vy], ics={vx.subs(t, 0):v0, vy.subs(t, 0):0})

sp.pprint([s.simplify() for s in sol])

plt.figure()
sp.plot(sol[0].subs(q, 1).subs(m, 1).subs(v0, 1).subs(E, 1).subs(B, 1).rhs,
        sol[1].subs(q, 1).subs(m, 1).subs(v0, 1).subs(E, 1).subs(B, 1).rhs)
```

```python
plt.figure()
sp.plot_parametric(sol[0].subs(q, 1).subs(m, 1).subs(v0, 1).subs(E, 1).subs(B, 1).rhs,
                   sol[1].subs(q, 1).subs(m, 1).subs(v0, 1).subs(E, 1).subs(B, 1).rhs)
```

```python
plt.figure()
sp.plot(sp.integrate(sol[0].subs(q, 1).subs(m, 1).subs(v0, 1).subs(E, 1).subs(B, 1), t).rhs,
        sp.integrate(sol[1].subs(q, 1).subs(m, 1).subs(v0, 1).subs(E, 1).subs(B, 1), t).rhs)
```
