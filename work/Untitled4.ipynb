Se tiene la funci´on:
    F (r) = Q**2 / r**2 si si r ≤ R,
    √Q / r si r > R,
    
siendo R y Q dos par´ametros, y r el m´odulo del vector de coordenadas (x, y).
Defina una funci´on Ffun(x, y, R, Q) que devuelva el valor de r y el del F correspondiente, que
sea v´alida para invocaciones en las que x e y sean arrays y que utilice la funci´on where de numpy.
(0.7 puntos)

```python
from numpy import *




def Ffun(x,y,R,Q):
    r=sqrt(x**2+y**2)
    F=where(r<R, Q**2/r**2, sqrt(Q)/r)
    return(r,F)
```

Prepare dos arrays de numpy (0.2 puntos):

 * El array x, que contiene las abscisas de los puntos donde queremos calcular F que est´an
espaciadas uniformemente entre x1 = R y x2 = 4R con nx = 101 puntos, siendo R = 20.0.

 * El array y, que contiene las ordenadas de los puntos donde queremos calcular F que est´an
espaciadas uniformemente con espaciado ∆y = 0.08R, partiendo de y1 = R hasta y2 = 4R (sin
incluir), siendo R = 20.0.

```python
R=20
x=linspace(R,4*R,101)
y=arange(R,4*R,0.08*R)
y.shape
```

Cambie la forma a los arrays anteriores de manera que puedan se usados en la funci´on Ffun. (0.2
puntos)

```python
x.shape=(101,1)
y.shape=(1,38)
```

Obtenga el array r y el F correspondiente llamando a la funci´on Ffun con los arrays x e y anterior-
mente creados y los siguientes datos: Q = 2.0, R = 20.0. (0.2 puntos)

```python
r,F=Ffun(x,y,R=20,Q=2)
```
