```python
from numpy import *
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
R = 1
v = 1
w = 1

def f(r, t):
    return array([
        r / R * cos(w * t) + v * t,
        - r / R * sin(w * t),
        v - v * r / R * sin(w * t),
        -v * r / R * cos(w * t)
    ])
```

```python
ts = arange(0 + 10/50, 10 + 10/50, 10/50).reshape([1, 50])
print(ts)
ts.shape
```

```python
rs = array([ 0.5, 0.75, 1 ]).reshape([3, 1])
print(rs)
rs.shape
```

```python
f(1, 0)
```

```python
f(rs[0][0], ts[0][0])
```

```python
d = f(rs, ts)

d.shape
```

```python
# shape has to be(3, 50, 4)
d = moveaxis(d, 0, -1)

d.shape
```

```python
f(rs, ts)
```
